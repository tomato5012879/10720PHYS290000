```python
import numpy as np
import matplotlib.pyplot as plt
import math

h = 6.62*10**-34
c = 3*10**8
k = 1.38*10**-23
num = 100

v = np.linspace(10**0, 10**15, 100, endpoint=True)
#print(v)
B = 2*h*v**3/c**2/(math.e**(h*v/k/5000)-1)

plt.plot(v, B)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Intensity Watts/Hz/m^2')

dB = np.random.normal(0., 10**-8.5, num)
B += dB

plt.plot(v,B)

plt.show()
```
