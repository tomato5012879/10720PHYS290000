```python
import numpy as np
import math
import matplotlib.pyplot as plt

pigs_x = np.random.random(1)*10
pigs_y = np.random.random(1)*10

v = 20
angle=math.pi/4
g = 9.8
Vx = v*math.cos(angle)
Vy = v*math.sin(angle)
t = np.linspace(0,3,20)

x = Vx*t
y = Vy*t - 1/2*g*(t**2)
plt.plot(x,y)

D = (x-pigs_x)**2 + (y-pigs_y)**2

if (D.all==0):
    print('打到!')
else :
    print('沒打到')

plt.plot(pigs_x,pigs_y, '.')
plt.xlabel('X')
plt.ylabel('y')
plt.xlim(0,50)
plt.ylim(0,20)
plt.show()
```
